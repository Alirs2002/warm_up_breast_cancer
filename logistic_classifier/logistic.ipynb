```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
all_data = load_breast_cancer()

x_data = all_data.data
y_data = all_data.target
```

```python
pd_data = pd.DataFrame(x_data)
pd_target = pd.DataFrame(y_data,columns=["target"])
```

```python
pd_data = pd.concat([pd_data,pd_target],axis=1)
pd_data
```

```python
train,test = train_test_split(pd_data,random_state=44,train_size=0.75)
```

```python
train.shape
```

```python
test.shape
```

```python
x_train = train.drop(columns="target")
```

```python
train
```

```python
x_train
```

```python
y_train = train["target"]
y_train
```

```python
x_test = test.drop(columns="target")
x_test
```

```python
y_test = test["target"]
y_test
```

```python
lr = LogisticRegression()

lr.fit(x_train,y_train)
```

```python
predicted_result = lr.predict(x_test)
predicted_result
```

```python
print(lr.score(x_test,y_test))
```
